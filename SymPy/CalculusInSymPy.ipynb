# Basic Calculus with SymPy

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

## Formula

```python
x, y, z = sp.symbols("x y z")
formula = x**5 - 4*x**4 - 10*x**3 + 40*x**2 + 9*x - 11
formula
f = sp.lambdify(x, formula)
x_vals = np.arange(-3.5,5,.1)
```

## Derivative

```python
fprime = sp.diff(formula, x)
fprime = sp.lambdify(x,fprime)
```

## Graph it

```python
current_point = (1,f(1))
```

```python
fig, ax = plt.subplots(1,1,figsize=(4,6))

#ax.suptitle("Slope of Tangent Point")
ax.set_ylim(-60,140)
ax.plot(x_vals,f(x_vals),label=r'$5 x^{4} - 16 x^{3} - 30x^{2} + 80 x + 9$')
ax.axline((0,0),slope=0.0,color="gray")
ax.axline((0,0),(0,1),color="gray")


ax.scatter(current_point[0],current_point[1])
ax.axline(current_point,slope=fprime(current_point[0]),color="r",linestyle="--",alpha=.5,label=str(current_point[0]))

plt.fill_between(x_vals,f(x_vals),0,where=(x_vals > -3) & (x_vals <= 2),color="red")

plt.legend()
plt.show()
                
```

## Integrate

```python
xmin = -3
xmax = 2
```

```python
sp.Integral(formula, (x,xmin,xmax))
```

```python
area = sp.integrate(formula, (x,-3,-1))
```
