```python
fig, ax = plt.subplots(1,2,figsize=(7,3),dpi=600)
fig.suptitle("Position of red and magenta dots on my computer screen, visualized", fontsize = 20, y=1.1)

ticks = np.linspace(0,1,5, dtype='float')


ax[0].set_title('red ones')
ax[0].scatter(x,y, color='red')
ax[0].set_xlabel("x values")
ax[0].set_ylabel("y values")
ax[0].set_xlim(0,1)
ax[0].set_ylim(0,1)

ax[1].set_xlabel("x values")
ax[1].scatter(x,y1, color='magenta')
ax[1].set_title('magenta ones')
ax[1].set_xlim(0,1)
ax[1].set_ylim(0,1)

plt.show()
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
y1 = np.random.rand(100)
y2 = np.random.rand(100)
y3 = np.random.rand(100)
```

```python
## make the above graph with a function

def dot_graph(ax, x, y, title="dot graph",dot_color="red"):
    ax.set_title(title)
    ax.set_xlabel("x values")
    ax.set_ylabel("y values")
    ax.set_xlim(0,1)
    ax.set_ylim(0,1)
    ax.scatter(x,y,color=dot_color, alpha=.5,edgecolor='maroon')
    return ax
```

```python
fig, ax = plt.subplots(4,4,figsize=(10,10),dpi=100,layout="constrained")
fig.suptitle("Matthew's Red Dot Collection", fontsize = 20, y=1.05)

ticks = np.linspace(0,1,5, dtype='float')

dot_graph(ax[0,0], x, y, 'red dots', "red")
dot_graph(ax[0,1], x, y, 'red dots', "salmon")
dot_graph(ax[1,0], x, y, 'red dots', "maroon")
dot_graph(ax[1,1], x, y, 'red dots', "crimson")
dot_graph(ax[0,2], x, y, 'red dots', "darkred")
dot_graph(ax[0,3], x, y, 'red dots', "tomato")
dot_graph(ax[1,2], x, y, 'red dots', "coral")
dot_graph(ax[1,3], x, y, 'red dots', "firebrick")
dot_graph(ax[2,0], x, y, 'red dots', "xkcd:wine red")
dot_graph(ax[2,1], x, y, 'red dots', "xkcd:blood red")
dot_graph(ax[2,2], x, y, 'red dots', "xkcd:deep red")
dot_graph(ax[2,3], x, y, 'red dots', "xkcd:orange red")
dot_graph(ax[3,0], x, y, 'red dots', "xkcd:burnt umber")
dot_graph(ax[3,1], x, y, 'red dots', "xkcd:vermillion")
dot_graph(ax[3,2], x, y, 'red dots', "xkcd:strawberry")
dot_graph(ax[3,3], x, y, 'red dots', "xkcd:fire engine red")
```

## Big O

```python
x = np.linspace(0,3,100)
```

```python
fig, ax = plt.subplots()

ax.set_title(r'Big $O$')
ax.plot(x,x,label="Linear")
ax.plot(x,x**2, label="Quadratic")
ax.plot(x,x**8, label="Octic")
ax.set_xlim(0,3)
ax.set_ylim(0,10)

ax.annotate(r"intersection of $\mathcal{O}(n)$ and $\mathcal{O}(n^2)$",xy=(1,1),xytext=(0,4),arrowprops=dict(facecolor="k",shrink=.05))

plt.legend()
plt.show()
```

## Bar Chart

```python
data = {
    "apples": np.random.normal(55,7,40),
    "grapes": np.random.normal(60,12,52),
    "pears": np.random.normal(50,20,36),
    "mangos": np.random.normal(67,40,61)
}
```

```python
fig, ax = plt.subplots(1,1,figsize=(4,5))
categories = data.keys()
means = [arr.mean() for arr in data.values()]
errs = [arr.std() for arr in data.values()]

ax.set_title("Matthew's Fruit")
ax.set_xlabel("these are the names of the fruits")
ax.set_ylabel("these are the number of each fruit matthew has, based on the look in his eye when asked about them")
ax.annotate("20",xy=(0,5),xytext=(-0.12,5),color="w")
ax.bar(categories, means,yerr=errs)

plt.show()
```
